```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3: # as want to take top 3 actors of movie
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split()) # as overview is string so we will change string to list
```

```python
movies.sample(5)
```

```python
# to remove spaces between the one sentence or name of one person otherwise it can cause problem to us
# like Sam Worthington and Sam Mendes these are two different person but when model take one word sam then model can get confuse and can give wrong prediction to us

def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
# movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

We have to make the website on which if user search for any movie by it's name then we also have to suggest him so other movies which is related to the movie he search like some action , same director , actor , means simply have to suggest the movie which have the same or we can say similar tags as that of the movie he search for

and now the question arise how we can get that two tags are similar to each other 

for this we will be using vectorization of text means converting the text of tag of movie into vectors , and whenever somebody search for the particular vector of tag then the nearest neighbour vector tags ka movies will be suggest to user

For this vectorization of text we will be using the method "out of bag" in this first we concat all the tags -> then we get the large text , then we count the frequency of each word in that text , then select the top like 500 words with maximum frequency

then from this top 5000 words we see like word1 comes in tag 1 how many times ? ,word2 comes in tag 1 how many times ?,.....,word5000 comes in tag 1 how many times ? 
similarly done for each tags, then we get the table from this of shape 4806x5000 and each record from this is act like the vector for each movie or movie tag

Before doing all this vectorization we have to first remove the stopwords

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tags'].apply(stem)
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') # remove english language stopwords and create the vector with 5000 words
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
feature_names = cv.get_feature_names_out()
print(feature_names[ :-1])
for name in feature_names:
    print(name +'\n')
len(feature_names)
```

```python
# it have some numerical features , which we think to remove or not to consider but there is person who want to see the movie of particular year so this can be helpful therefore we will not remove the numerical values 
# but there is also one problem that there are features like [action,actions] which are both same so it can cause problem while recommending the movie to user as user search for movie have the tag action so our model not recommend him the movie which have tags like actions therefore to correct this we will use nltk which convert this to [action,action]
```

```python
# so before running count vectorizstion we have to run nltk code first
```

```python
# after all this , we have 4806 records or vectors with 5000 featurs or dimension of vectors
# now we have to calculate the distance between the vectors of two movies , more the distance less the similarity between the coressponding movies (of vector)
# in this we will calculate cosine distnace means angle between the two vectors instead of eculiean distance 
# since whenever you work for large dimensionality eucliean distance fails 
```

```python
from sklearn.metrics.pairwise import cosine_similarity # this will give us the output as the similarity wrt to cosine distance
# the cosine distance is inversely proportional to similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]: # take the top 5 movies which similar to movie search by user
        print(new.iloc[i[0]].title)
        
    
```

```python
recommend('Gandhi')
```
